```python
# ### Assignment 1: Handling Division by Zero

# # Write a function that takes two integers as input and returns their division. Use try, except, and finally blocks to handle division by zero and print an appropriate message.
def divide_numbers(num1, num2):
    try:
        num = num1/num2
    except ZeroDivisionError as  ZDE :
        print(f"number can't be divided by Zero {ZDE}")
    finally:
        print(f"execution completed")
    return num

divide_numbers(20, 2)
```

```python

# ### Assignment 2: File Reading with Exception Handling

# Write a function that reads the contents of a file named `data.txt`. Use try, except, and finally blocks to handle file not found errors and ensure the file is properly closed.
def read_file(file_name):
    try :
        with open(file_name,'r') as file:
            print(file.read())
    except FileNotFoundError as fn:
        print(f"file not found {fn}")
    finally :
        try :
            file.close()
        except NameError  as ne:
            print(f"file not {ne}")

read_file("sample.txt")
```

```python
# ### Assignment 3: Handling Multiple Exceptions

# Write a function that takes a list of integers and returns their sum. Use try, except, and finally blocks to handle TypeError if a non-integer value is encountered and print an appropriate message.
def integersum(lst):
    total =0
    try :
        for i in lst:
            total += i
    except TypeError as ty:
        print(f"this is non integer value {ty}")
        total = 0
    finally :
        return total

integersum([1,2,3,4,5,6,"ayush"])
```

```python

# ### Assignment 6: Nested Exception Handling

# Write a function that performs nested exception handling. It should first attempt to convert a string to an integer, and then attempt to divide by that integer. 
# Use nested try, except, and finally blocks to handle ValueError and ZeroDivisionError and print appropriate messages.
from pandas import notna


def nestedException(name):
    try :
        try :
            name1= int(name)
        except ValueError as ve:
            print(f"value error {ve}")
        finally :
            print(f"check converting string to integer")
        if name1 is not None:
            try :
                result = 10 /name1
            except ZeroDivisionError  as ze:
                print(f"zero division error {ze}")
                result = None
            finally :
                print("division  attempt  completed.")
                return result
    finally:
        print("overall exception completed")

print(nestedException('0'))  # None
print(nestedException('a'))  # None
print(nestedException('2'))  # 5.0
```

```python

# ### Assignment 8: Exception Handling in Network Operations

# Write a function that attempts to open a URL and read its contents. Use try, except, and finally blocks to handle network-related errors and print an appropriate message.
# from flask import request
import requests
def openURL(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        return response.text
    except requests.RequestException as r:
        print(f"network error {r}")
        return None
    finally :
        print('execution  completed')

print(openURL('https://jsonplaceholder.typicode.com/posts/1'))
print(openURL('https://nonexistent.url'))
```

```python

# ### Assignment 9: Exception Handling in JSON Parsing

# Write a function that attempts to parse a JSON string. Use try, except, and finally blocks to handle JSONDecodeError if the string is not a valid JSON and print an appropriate message.
import json
def jsonParsing(json_string):
    try :
        result = json.loads(json_string)
        # return result
    except json.JSONDecodeError as e:
        print(f"error message : {e}")
        return None
    finally :
        print("Execution complete.")
    
print(jsonParsing('{"name": "John", "age": 30}'))  # {'name': 'John', 'age': 30}
print(jsonParsing('Invalid JSON'))  # None
```

```python


# ### Assignment 10: Custom Exception Handling

# Define a custom exception named `NegativeNumberError`. Write a function that raises this exception if a negative number is encountered in a list. 
# Use try, except, and finally blocks to handle the custom exception and print an appropriate message.

from operator import ne


class NegativeNumberError(Exception):
    pass

def check_for_negative(lst):
    try:
        for num in lst:
            if num < 0:
                raise NegativeNumberError("negative number exception")
    except NegativeNumberError as ne :
        print(f"error : {ne}")
    finally :
        print("execution message")

check_for_negative([1,2,3,4,5,5])
check_for_negative([-1])
```
