```python
# 09 Write a function that takes a 3x3 matrix (nested list) as input and returns its transpose. Print the original and transposed matrices.
def transposefunction(list):
    
    transpose = [[list[i][j] for i in range(len(list)) ] for j in range(len(list[0]))]
    return transpose
list= [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]
transpose= transposefunction(list)
# for tran in transpose:
print(transpose)
for lit  in list:
    print(lit)
```

```python
# 10 Write a function that takes a nested list and flattens it into a single list. Print the original and flattened lists.

def function(list):
    transpose = [item for lst in list for item in lst]
    return transpose

list= [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]
res= function(list)
print(res)
```

```python
# 11 Create a list of the first 10 positive integers. Remove the elements at indices 2, 4, and 6, and insert the element '99' at index 5. Print the modified list.
list = [i for i in range(1,11)]
del list[2]
del list[4]
del list[6]
print(list)
list.insert(5, 99)
print(list)
```

```python
#12 Create two lists of the same length. Use the `zip` function to combine these lists into a list of tuples and print the result.
def ZipFunction(list1 , list2):
    return list(zip(list1, list2))
    
list1 = [1,3,4,5,6]
list2 = [1,2,3,4,5]
listoftuple= ZipFunction(list1, list2)
print(listoftuple)
```

```python
#13 Write a function that takes a list and returns a new list with the elements in reverse order. Print the original and reversed lists.
def reverselist(list):
    return list[::-1]

list = [1,2,3,4,5]
print(f"original list : {list}")
reverse_list = reverselist(list)
print(f"reverse_list : {reverse_list}")
```

```python
#14 Write a function that rotates a list by n positions. Print the original and rotated lists.

list = [i for i in range(6)]
print(f"original list :{list}")
list.reverse()
rotatedlist= [ i for i in list]
print(rotatedlist)
```

```python
#15 Write a function that takes two lists and returns a new list containing only the elements that are present in both lists. Print the intersected list.
def intersectionList(list, list1):
    return [lst for lst in list if lst in list1]

list = [1,2,4]
list1 = [1,2,3]
ret= intersectionList(list, list1)
print(ret)
```
