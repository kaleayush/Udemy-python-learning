the filter() function constructs an iterator from elements of an iterable  for which a function return  true . it is used to filter out item  from list based on condition.

```python
def even(n):
    if n%2==0:
        return True

even(4)
```

```python
list1 =[x for x in range(1,11)]
list(filter(even, list1))
```

```python
list(filter(lambda x : x>5 , list1))
```

```python
# filter with a lambda function  and multiple condition
list(filter(lambda x : x>5 and x%2==0 , list1))
```

```python
list(filter(lambda x :  x%2==0 , list1))
```
