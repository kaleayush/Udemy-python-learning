```python
#positional arguments
def manyargument(*args):
    print(type(args))
    for arg in args:
        print(arg, end=" ")
manyargument(1,2,3,4,5,56,6,"ayush")
```

```python
#keyword argumnets
def function(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

function(name= "ayush", age = 25, address= "indore")
```

```python
#keyword and positional arguments

def function(*args, **Kwargs):
    for key, value in Kwargs.items():
        print(f"{key}: {value}")
    for arg in args:
        print(arg)

function(1,2,3,4,4,5,name= "ayush", age = 25, address= "indore")
```
