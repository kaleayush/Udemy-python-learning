#Decorator are the powerfull feature in python that allows to change the behavior of function or class and method. they are commonly used  to add functionality  to function or methods without  modifiying their actual code this lession covers the basic of decorators inclluding how create and use them

```python
#function copy 

def Welcome():
    return "Wel-come to advance python"
```

```python
wel = Welcome # creating copy of welcome function 
print(wel())
del Welcome
print(wel())
```

```python
#closures is a method within a method 

def welcome():
    message= "this is message"
    def sub_welcome():
        print(f"this is the welcome sub method {message}")
    return sub_welcome()

welcome()
```

```python
def welcome(func):
    message= "this is message"
    def sub_welcome():
        print(f"this is the welcome sub method {message}")
        func("this is function")
        print(f"this is the welcome sub method {message}")

    return sub_welcome()

welcome(print)
```

```python
def welcome(func , lst):
    message= "this is message"
    def sub_welcome():
        print(f"this is the welcome sub method {message}")
        print(func(lst))
        print(f"this is the welcome sub method {message}")

    return sub_welcome()

welcome(len , [1,2,3,4])
```

```python
def welcome(func):
    message= "this is message"
    def sub_welcome():
        print(f"this is the welcome sub method {message}")
        func()
        print(f"this is the welcome sub method {message}")

    return sub_welcome()
```

```python
def course():
    print("this is testing ")
```

```python
welcome(course)
```

```python
@welcome #this is the decorators
def course():
    print("this is testing ")
```

```python
## decorators

def my_decorator(func):
    def wrapper():
        print("something is happing before the function is called")
        func()
        print("something is happing before the function is called")
    return wrapper
```

```python
@my_decorator
def say_hello():
    print("hello")
```

```python
say_hello()
```

```python
#decorators with arugments

def repeat(n):
    def decorator(func):
        def wrapper(*args , **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator
```

```python
@repeat(3)
def say_hello():
    print("hello")
```

```python
say_hello()
```
