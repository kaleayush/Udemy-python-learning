Dictionaries are unorder collection of items. they store item in key value pair. key must be unique and immutable(eg string , number or tuple) while value can be of any type


```python
# creation of dictionarie
my_dic= {}
print(type(my_dic))

empty_dic= dict()
print(type(empty_dic))
```

```python
student = {1:"ayush", 2 : "rounak" , 3: 26}
print(student)
```

```python
student = {1:"ayush", 2 : "rounak" , 3: 26 , 1:"pune" , "pune" : "city name"}
print(student)
```

```python
#accessing dictionarie element 
print(student["pune"])
print(student.get("pune"))
print(student.get("lastName"))
```

```python
#modifing element
#updated value of key 
student[1] = "aman"
print(student)
#added new key value 
student["address"] ="inodre mp"
print(student)
```

```python
#delete key from dictionary   
del student["pune"]
print(student)
```

```python
#dictionary methods
key = student.keys()
print(key)
value = student.values()
print(value)

items= student.items()
print(items)
```

```python
#shellow copy 
student2 = student
print(student)
print(student2)
```

```python
student[1] ="rounak"
print(student)
print(student2)
```

```python
# copy
student1 = student2.copy()
print(student1, student)
student[1]= "aman"
print(student1, student)
```

```python
#iteration over dictionary
#iteration can be done on key , value and items
for key in student.keys():
    print(key)

for value in student.values():
    print(value)

for key, value in student.items():
    print(f"{key} : {value}")
```

```python
#nested dictionary 
students = {"student1" : {"name" :"ayush" , "age" : 25 , "address" : "Indore"} , "student2" : {"name" :"aman" , "age" : 25 , "address" : "Indore"}}
print(students)
```

```python
#iterating of nested dict
for st_key , st_value in students.items():
    for key , value in st_value.items():
        print(value)
```

```python
#dic comphehension
sq= {x: x**2 for x in range(10) if x%2 ==0} 
print(sq)
```

```python
list= [1,2,2,3,3,3,4,4,4,4]
dict= {}
for num in list:
    if num in dict:
        dict[num] = dict[num]+1
    else :
        dict[num] = 1
print(dict)
```

```python
#merge  2 dict
merge = {**dict , **sq}
print(merge)
```
