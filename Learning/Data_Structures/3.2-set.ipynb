Set is a built in data type in python used to store collection of unique items. they are unorderd meaning element do not follow specific order and they do not allow duplicate element. set are usefull  for membership tests, eliminating  duplicate entries, and performing  mathematical set of operation like union , intersection ,  difference, and symmetric difference.

```python
#creation of set
my_set ={1,2,34,5}
print(my_set)
print(type(my_set))
my_dic = {}
print(type(my_dic))
```

```python
clas_set = set()
print(type(clas_set))
```

```python
list_to_set = set([1,2,3,4,5,5])
print(list_to_set)
```

```python
#Basic set operation 
# to add element in set
list_to_set.add(6)
print(list_to_set)
list_to_set.add(6)
print(list_to_set)
```

```python
# how to remove element from set 
list_to_set.remove(5)
print(list_to_set)
```

```python
list_to_set.discard(5)
print(list_to_set)
```

```python
ele= list_to_set.pop()
print(ele)
```

```python
list_to_set.clear()
print(list_to_set)
```

```python
#set membership test
my_set= {1,2,4,5,3,5,6,7}
print(3 in my_set)
print(10 in my_set)
```

```python
#set mathematic operation 
set1= {1,2,3,4,5,6}
set2= {4,5,6,7,8}

print(set1.union(set2))

print(set1.intersection(set2))
# set1.intersection_update(set2)
print(set1)
print(set1.difference(set2))
print(set1.symmetric_difference(set2))
```

```python
set1 ={1,2,3}
set2= {3,4,5}
print(set1.issubset(set2))
```
