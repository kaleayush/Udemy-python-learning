```python
## read the whole file
with open(file="example.text" , mode ="r" )  as file:
    content = file.read()
    print(content)
```

```python
## read the file line by line
with open(file= "example.text" , mode= "r") as file:
    for line in file:
        print(line.strip())
#strip remove new empty line after line
```

```python
#writing the file (overwriting)
with open(file="example.text" ,mode= "w") as file:
    file.write("Hello world\n")
    file.write("this is new line\n")
```

```python
# writing the file (without overwriting)
with open (file ="example.text" , mode= "a") as file:
    file.write("Hello world 222\n")
    file.write("this is new line222")
```

```python
# writing a list of line
lines="line \n line 2 \n line 3"
with open(file="example.text", mode="a") as file1:
    file1.writelines(lines)
```

```python
# read binary file     
with open(file="example.bin", mode="rb") as file:
    content =file.read()
    print(content)
```

```python
# write binary files
data = b'\x00\x01\x02\x03'
with open(file="example.bin" , mode= "wb") as file:
    file.write(data)
```

```python
#read the content of example.text file and put in distination.text file
with open(file="example.text", mode="r") as file:
    content = file.read()

with open(file="destination.text", mode="w") as file1:
    file1.writelines(content)
```

```python
def count_word(file_path):
    with open(file=file_path , mode="r") as file:
        lines = file.readlines()
        print(lines)
        print(type(lines))
        line_count= len(lines)
        word_count= sum(len(line.split()) for line in lines)
        char_line = sum(len(line) for line in lines)
        print(line_count)
        print(char_line)
    return line_count , word_count, char_line
file ="example.text"
line, word , char= count_word(file)

print(line, word , char)
```
