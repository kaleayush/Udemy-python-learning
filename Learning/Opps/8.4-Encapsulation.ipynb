Encapsulation and Abstraction

Encapsulation and Abstraction are two fundamental principle of oops that helps in designing robust manintainable and reusable code .Encapsulation involves bundling data and methods that operate on the data within a single unit while abstration involves hiding complex implementation detail and exposing only the necessary features.

```python
## Encapsultion with Getter and setter methods
#public protected ,private , 


class Persion:
    def __init__(self, name , age , address , number):
        self.name = name #public variable 
        self.age = age #public variable 
        self.__address= address #private variable 
        self._Mobilenumber = number
#private variable is not access from out side the class
#proteted variable is not access from out side the class but access through base class
    def getattr(self):
        return self.__address


class Emp(Persion):
    def __init__(self, name, age, address, number):
        super().__init__(name, age, address, number)

persion = Emp("ayush" ,25, "indore" ,1233)
# persion.getattr()

persion._Mobilenumber
```
