```python
class Bank:
    def __init__(self , name , balance):
        self.name = name
        self.balance = balance
    
    def addAmount(self , amount):
        self.balance += amount
        print("amount is added to your account ")

    def withdrawAmount(self, amount):
        if self.balance <= 0 :
            print("No amount in account")
        elif 0 <= (self.balance - amount) :
            print("insufficient amount")
        else:
            self.balance-= amount
    
    def getCurrentAmount(self):
        print(self.balance)


rounak = Bank("rounak" , 500)
rounak.getCurrentAmount()
        
```
