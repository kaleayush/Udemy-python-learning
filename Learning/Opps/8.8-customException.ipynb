Custom except 

```python
class Error(Exception):
    pass

class dobException(Error):
    pass
```

```python
year = int(input("enter the DOB"))
age = 2026 - year
try :
    if age<= 30 and age >= 20:
        print(f"the age is valid so you can apply for exam")
    else :
        raise dobException 
except dobException :
    print("sorry your age should be grater than 20and less than 30")
```
