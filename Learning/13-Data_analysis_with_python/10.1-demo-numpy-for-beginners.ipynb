# NumPy Demo For Beginners

Think of NumPy like a very fast calculator for lists of numbers.

A normal Python list is like a row of school notebooks. NumPy arrays are like a neat math table where the computer can do many sums at once.

In real apps, NumPy is used for data analysis, machine learning, image processing, finance, games, and scientific calculations.

## 1. Import NumPy

`np` is a short nickname. Programmers use it so they do not need to type `numpy` again and again.

```python
import numpy as np

print("NumPy is ready")
```

## 2. Create Arrays

An array is like a box that stores numbers in a neat line or table.

```python
# From a Python list
marks = np.array([80, 90, 75, 60, 95])
print("Marks:", marks)

# 2D array, like rows and columns in a classroom seating chart
class_marks = np.array([
    [80, 90, 75],
    [60, 95, 88]
])
print("Class marks:\n", class_marks)
```

## 3. Common Array Builders

These methods make ready-made arrays. They are useful when apps need empty tables, default values, or number ranges.

```python
print("zeros:", np.zeros(5))          # five 0 values
print("ones:", np.ones(5))            # five 1 values
print("full:", np.full(5, 7))         # five 7 values
print("arange:", np.arange(1, 11))    # numbers from 1 to 10
print("linspace:", np.linspace(0, 1, 5)) # 5 equal steps from 0 to 1
print("eye:\n", np.eye(3))            # identity matrix, used in math and ML
```

## 4. Array Information

Before using data, apps often ask: How big is it? What type is it? How many dimensions does it have?

```python
data = np.array([[10, 20, 30], [40, 50, 60]])

print("Array:\n", data)
print("shape:", data.shape)   # rows and columns
print("ndim:", data.ndim)     # number of dimensions
print("size:", data.size)     # total items
print("dtype:", data.dtype)   # data type
```

## 5. Change Shape

`reshape` is like rearranging chocolates from one long row into a box with rows and columns. The number of chocolates must stay the same.

```python
numbers = np.arange(1, 13)
print("Original:", numbers)

table = numbers.reshape(3, 4)
print("3 rows and 4 columns:\n", table)

print("flatten:", table.flatten()) # makes a new 1D copy
print("ravel:", table.ravel())     # makes a 1D view when possible
```

## 6. Pick Items With Indexing And Slicing

Index means position. Python starts counting from 0, so the first item is at position 0.

```python
scores = np.array([45, 60, 72, 88, 91, 30])

print("First score:", scores[0])
print("Last score:", scores[-1])
print("From index 1 to 3:", scores[1:4])
print("Every second score:", scores[::2])
```

```python
matrix = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])

print("Item at row 1, column 2:", matrix[1, 2])
print("First row:", matrix[0, :])
print("Second column:", matrix[:, 1])
```

## 7. Boolean Masking

A mask is like a filter. It keeps only the values that pass a condition.

```python
scores = np.array([45, 60, 72, 88, 91, 30])

passed = scores >= 50
print("True means pass:", passed)
print("Only passed scores:", scores[passed])
print("Same thing shorter:", scores[scores >= 50])
```

## 8. Fancy Indexing

Fancy indexing means choosing many positions at once.

```python
players = np.array(["Asha", "Ravi", "Neha", "Kabir"])
chosen_positions = [0, 2]

print(players[chosen_positions])
```

## 9. Math On Whole Arrays

NumPy can do math on every item at once. This is why it is fast and popular.

```python
prices = np.array([100, 200, 300])

print("Add 10:", prices + 10)
print("Subtract 20:", prices - 20)
print("Double:", prices * 2)
print("Half:", prices / 2)
print("Square:", prices ** 2)
```

```python
values = np.array([-4, -1, 0, 9, 16])

print("abs:", np.abs(values))
print("sqrt:", np.sqrt(np.array([4, 9, 16])))
print("round:", np.round(np.array([1.234, 5.678]), 2))
print("exp:", np.exp(np.array([1, 2, 3])))
print("log:", np.log(np.array([1, np.e, np.e**2])))
```

## 10. Aggregation Methods

Aggregation means making many numbers into one answer, like total, average, smallest, or biggest.

```python
marks = np.array([80, 90, 75, 60, 95])

print("sum:", np.sum(marks))
print("mean:", np.mean(marks))      # average
print("median:", np.median(marks))  # middle value
print("min:", np.min(marks))
print("max:", np.max(marks))
print("std:", np.std(marks))        # how spread out values are
print("var:", np.var(marks))        # another spread measurement
print("argmin:", np.argmin(marks))  # position of smallest value
print("argmax:", np.argmax(marks))  # position of biggest value
print("cumsum:", np.cumsum(marks))  # running total
```

## 11. Axis: Row Wise And Column Wise

Axis tells NumPy the direction.

- `axis=0` means go down each column.
- `axis=1` means go across each row.

```python
marks_table = np.array([
    [80, 90, 70],
    [60, 95, 85],
    [75, 65, 88]
])

print("Table:\n", marks_table)
print("Column total, subject wise:", np.sum(marks_table, axis=0))
print("Row total, student wise:", np.sum(marks_table, axis=1))
print("Student averages:", np.mean(marks_table, axis=1))
```

## 12. Broadcasting

Broadcasting means NumPy stretches a small value or small array to match a bigger array.

Example: Add bonus marks to every student.

```python
marks_table = np.array([
    [80, 90, 70],
    [60, 95, 85],
    [75, 65, 88]
])

bonus_per_subject = np.array([5, 2, 3])
print(marks_table + bonus_per_subject)
```

## 13. `where`: If Else For Arrays

`np.where` chooses values based on a condition. It is like saying: if pass, write Pass, else write Fail.

```python
scores = np.array([45, 60, 72, 88, 91, 30])

result = np.where(scores >= 50, "Pass", "Fail")
print(result)

# Give grace marks only to students below 50
new_scores = np.where(scores < 50, scores + 5, scores)
print(new_scores)
```

## 14. Sorting And Unique Values

Apps often need to sort values or remove duplicates.

```python
items = np.array([3, 1, 2, 3, 5, 2, 1])

print("sort:", np.sort(items))
print("unique:", np.unique(items))
print("argsort:", np.argsort(items)) # positions that would sort the array
```

## 15. Join And Split Arrays

Joining means putting arrays together. Splitting means cutting one array into parts.

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("concatenate:", np.concatenate([a, b]))
print("stack:\n", np.stack([a, b]))
print("vstack:\n", np.vstack([a, b]))
print("hstack:", np.hstack([a, b]))
print("split:", np.split(np.array([1, 2, 3, 4, 5, 6]), 3))
```

## 16. Copy And View

A copy is a separate duplicate. A view is like another window looking at the same data.

This matters because changing a view can change the original array.

```python
original = np.array([10, 20, 30, 40])

copy_array = original.copy()
view_array = original[1:3]

copy_array[0] = 999
view_array[0] = 777

print("original:", original)
print("copy_array:", copy_array)
print("view_array:", view_array)
```

## 17. Change Data Type

Apps sometimes need numbers as integers, decimals, or text. `astype` changes the type.

```python
decimal_numbers = np.array([1.2, 2.8, 3.5])

print(decimal_numbers.astype(int))
print(decimal_numbers.astype(str))
```

## 18. Clip Values

`clip` keeps values inside a minimum and maximum limit.

Example: In an app, rating should stay between 1 and 5.

```python
ratings = np.array([0, 1, 3, 5, 8])
safe_ratings = np.clip(ratings, 1, 5)

print(safe_ratings)
```

## 19. Missing Values: NaN

`NaN` means Not a Number. It is used when data is missing or invalid.

```python
temperature = np.array([30, 32, np.nan, 29, 31])

print("isnan:", np.isnan(temperature))
print("mean gives nan:", np.mean(temperature))
print("nanmean skips missing value:", np.nanmean(temperature))
print("replace nan with 0:", np.nan_to_num(temperature, nan=0))
```

## 20. Random Numbers

Random numbers are used in games, testing, simulations, and machine learning.

Using a seed makes the random answer repeatable, so your teacher and your computer can see the same result.

```python
rng = np.random.default_rng(seed=42)

print("random decimals:", rng.random(5))
print("random integers:", rng.integers(1, 7, size=10)) # like rolling dice
print("normal data:", rng.normal(loc=50, scale=10, size=5))
print("choice:", rng.choice(["red", "blue", "green"], size=4))

cards = np.array([1, 2, 3, 4, 5])
rng.shuffle(cards)
print("shuffle changes original:", cards)
print("permutation returns shuffled copy:", rng.permutation([1, 2, 3, 4, 5]))
```

## 21. Matrix Math And Linear Algebra

This is used in machine learning, graphics, engineering, and recommendation systems.

Do not worry if it feels big now. First remember that matrices are number tables.

```python
x = np.array([[1, 2], [3, 4]])
y = np.array([[5, 6], [7, 8]])

print("dot:\n", np.dot(x, y))
print("matmul using @:\n", x @ y)
print("transpose:\n", x.T)
print("determinant:", np.linalg.det(x))
print("inverse:\n", np.linalg.inv(x))
```

```python
# Solve equations:
# 2a + b = 8
# a + 3b = 13

A = np.array([[2, 1], [1, 3]])
B = np.array([8, 13])

answer = np.linalg.solve(A, B)
print("a and b:", answer)
```

## 22. Save And Load NumPy Data

Apps save data so they can use it later.

```python
small_data = np.array([10, 20, 30])

np.save("small_data.npy", small_data)
loaded_data = np.load("small_data.npy")

print(loaded_data)
```

## 23. Mini App Example: Student Report

Now let us use many NumPy methods together.

Imagine an app that checks student marks and creates a small report.

```python
students = np.array(["Asha", "Ravi", "Neha", "Kabir"])
marks = np.array([
    [80, 90, 70],
    [45, 55, 60],
    [95, 88, 92],
    [30, 40, 35]
])

average_marks = np.mean(marks, axis=1)
result = np.where(average_marks >= 50, "Pass", "Fail")
top_student_index = np.argmax(average_marks)

print("Students:", students)
print("Average marks:", np.round(average_marks, 2))
print("Result:", result)
print("Top student:", students[top_student_index])
print("Top average:", average_marks[top_student_index])
```
