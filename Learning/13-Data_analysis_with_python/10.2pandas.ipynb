Pandas is a powerfull data manipulation library in python widely used for data analysis and data cleaning it provides two primary data structures : series and Dataframe. A series is a one dimensional arrey like object while a Dataframe is a two-dimensional, size-mutable and potentially heterogeneous tabular data structure with labeled axes (rows and columns).

```python
import pandas as pd
```

```python
## series 
# pandas is a one-dimension array like object 

import pandas as pd
data =  [1,2,3,4,5,6]
series = pd.Series(data)
print(series)
print(type(series))
```

```python
#create series from dictionary
data = {'a': 1 , 'b':2}
series_dic = pd.Series(data)
print(series_dic)
```

```python
data = [10,20,30]
index = ['a','b','c']
pd.Series(data , index= index)
```

```python
#Data frame
## ceate aData frame from dict of list
data = { 
    'name' : ['krish', 'John','jack' ],
    "age" : [25,30,45],
    "city":['inodre','new york','florida']
}
df = pd.DataFrame(data=data)
print(df)
print(type(df))
```

```python
#create a datafrane from lst of dict 

data = [
    {'name' : 'krish', 'age' : 25 , 'city': 'banglore'},
    {'name' : 'John', 'age' : 25 , 'city': 'banglore'},
    {'name' : 'Aman', 'age' : 25 , 'city': 'banglore'},
    {'name' : 'Anand', 'age' : 25 , 'city': 'banglore'},
]
df = pd.DataFrame(data=data)
print(df)
print(type(df))
```
