```python
import sqlite3
```

```python
connection = sqlite3.connect('example.db')
cursor = connection.cursor()
```

```python
cursor.execute('''
    create table if not exists User 
            (
               Id integer primary key,
               UserName string,
               profilepic string,
               phone_no integer
               )

''')

connection.commit()
```

```python
cursor.execute('''
    Insert into user (UserName, profilepic, phone_no) values ("ayush", "image.jpeg", 987654321),
               ("ayush01", "image.jpeg", 987654321),
               ("ayush02", "image.jpeg", 987654321);

''')

connection.commit()
```

```python
data = cursor.execute('''
    select * from user
''')

for i in data:
    print(i)
```

```python
cursor.execute('''
    select * from user
''')
data = cursor.fetchall()

for i in data:
    print(i)
```

```python
cursor.execute(
'''
    Create table sales (
        salesId integer primary key,
        sales_item string ,
        price integer
    )
'''
)
connection.commit()
```

```python
data = [
    ("clutchplate", 1000),
    ("clutchplate1", 1000),
    ("clutchplate2", 1000),
    ("clutchplate3", 1000),
]

cursor.executemany('''
    insert into sales (sales_item, price) values (?,?)
''', data)

connection.commit()


```

```python
cursor.execute('''
    select * from sales
''')

for i in data:
    print(i)
```
