map functon applies on given function to all items in an input list and return map object an iterator , this is usefull for transforming  data in list comphresion 

```python
def square(x):
    return x*x

square(5)
```

```python
list1= [x for x in range(1,11)]
list(map(square, list1))
```

```python
lst1 = [1,2,3] 
lst2 = [4,5,6]

list3 =list(map(lambda x, y: x+y , lst1 , lst2))
print(list3)
```

```python
# convert list of string no to int value using map
list1= ['1','2','3','4']
new_list=list(map(int, list1))
print(new_list)
```

```python
words=['apple','banan', 'chreey']
new_list_word= list(map(str.upper, words))
print(new_list_word)
```

```python
def returnName(data):
    return data["name"]

student= [
        {"name" :"aysuh", "age" :26},
        {"name":"aysuh", "age" :26}
        ]
list(map(returnName, student))
```
