```python
#temp conversion
def tempconversion(temp , unit):
    if unit == "c":
        return  temp * 9/5 +32
    elif unit == "F":
        return (temp-32)*5/9
    else :
        return None
tempconversion(32 ,"c")
tempconversion(89.6,"F")
```

```python
#password checking is strong or not 

def checkPassword(password):
    if len(password)<8:
        return False
    if not any(letter.isdigit() for letter in password):
        return False
    if not any(letter.islower() for letter in password):
        return False
    if not any(letter.isupper() for letter in password):
        return False
    else:
        return True

print(checkPassword("Akdsdflds"))
        
```

```python
#calculate total cost of item in shoping list
def calculateAmount(shoping):
    totalAmount = 0
    for item in shoping:
        totalAmount += item['amount'] * item['quantity']
    return totalAmount 

shoping =[
    {"product" : "shirt" , "amount" : 100 , "quantity" : 3},
    {"product" : "T-shirt" , "amount" : 90 , "quantity" : 5},
    {"product" : "jeans" , "amount" : 300 , "quantity" : 3},
    ]
print(calculateAmount(shoping))


```

```python
#check if string is palindrome or not   
def CheckPalindrome(str):
    str=str.lower().replace(" ","")
    return str==str[::-1]
print(CheckPalindrome("ABA"))
print(CheckPalindrome("Hello"))

```

```python
# calculate the factorial of a number using recursion 

def fact(n):
    if n ==0 :
        return 1
    else:
        return n * fact(n-1)

print(fact(5))
```

```python
# function to read file and count the frequrency of  each word
def frequencycount(file_path):
    wordCount= {}
    with open(file_path,'r' ) as file:
        for line in file :
            words = line.split()
            for word in words:
                word = word.lower().strip('.,!?;:"\'')
                wordCount[word] = wordCount.get(word, 0) +1
    return wordCount
file_path ="test.txt"
word_frequency = frequencycount(file_path)
print(word_frequency)
```
