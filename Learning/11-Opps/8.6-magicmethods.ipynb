Megic methods in py also known as dunder method (double underscore methods) are special methods that start end end with double underscores these methods enable you to define the behavior of object for built in operations, such as arithmetic operations, comparisions and more


magic method are predefined method on py that you can override to chnge the behavior of your objects. some common magic method include:

__init__ intialize a new instance of a class.
__str__ return a string representation of an objects.
__repr__ returns and offical string representation of an object.
__len__ return the length if an object.
__getitem__ gets an item from a container.
__setitem__ sets an item in a container.

```python
#Basic meagic method 
class Persion:
    def __init__(self, name , age):
        self.name = name
        self.age= age

p= Persion("ayush" , 25)
print(p)
```

```python
class Persion:
    def __init__(self, name , age):
        self.name = name
        self.age= age

    def __str__(self):
        return f"{self.name}, {self.age} year old"
    
    def __repr__(self):
        return f"persion(name ={self.name})"
p= Persion("ayush" , 25)
print(p)
print(repr(p))
```
