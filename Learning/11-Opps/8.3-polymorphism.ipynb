Polymorphism is a core concept in oops that allows objects of different classes to be treated as objects of a common superclass . it provides a way to perform a sigle action in different forms polymorphism is typically achived through method overriding and interfaces.

Method overriding 

method overriding allows a child class to provide a specific implementation of a method that is already defined in its parent class. 

```python
## this is method overiding .
## Base Class52
class Animal :
    def Speak(self):
        return "sound if the animal"

#drive class
class Dog(Animal):
    def Speak(self):
        return "woof"

class Cat(Animal):
    def Speak(self):
        return "meow!"

dog= Dog()
print(dog.Speak())
cat= Cat()
print(cat.Speak())
```

```python
## polymorphism with method and functions
class Shape :
    def Area():
        return "the area of the figure"
    
class Rectange(Shape):
    def __init__(self, width , height):
        super().__init__()
        self.width = width
        self.height = height

    def Area(self):
        return self.width * self.height
    
class Circle(Shape):
    def __init__(self, radius):
        super().__init__()
        self.Radius = radius

    def Area(self):
        return 3.14 * self.Radius * self.Radius
    

def print_Area(shape):
    print(f"are of shape is {shape.Area()}")

rectange = Rectange(4,5)
circle = Circle(4)

print_Area(rectange)
print_Area(circle)
```

Abstract base class are used to define common methods for a group of related object they can enforce that drived classed implement particular methods promting consistency across different implementations

```python
from abc import ABC , abstractmethod

#define abstract class
class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        return "car engine started"
    
class Motorcycle(Vehicle):
    def start_engine(self):
        return "motorcycle engine started"


def start_engine(vehicle):
    print(vehicle.start_engine())
## create objects of car and motorcycle 
car=Car()
motorcycle= Motorcycle()

start_engine(car)
```
