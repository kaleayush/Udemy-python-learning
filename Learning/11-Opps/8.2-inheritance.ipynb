Inheritance is a concept of oops in this child class inherite the property and method of parent class

```python
class parent :
    def __init__(self, cast, amount):
        self.cast = cast
        self.amount = amount 

class child(parent) :
    def __init__(self, cast, amount, name):
        super().__init__(cast, amount)
        self.name = name

Ayush = child("obc" , 5000 , "yush")
print(Ayush.cast)
```

Multiple inheritance
when a class inherits more than one class 

```python
class Car :
    def __init__(self , window, door, carType):
        self.window = window
        self.door = door
        self.carType  = carType
    def drive(self):
        print(f"this is car {self.carType}")
        
audi = Car(4, 4, "cng")
audi.drive()
```

```python
class Tesla(Car):
    def __init__(self, window , door , carType , is_SelfDrive):
        super().__init__(window, door, carType)
        self.is_SelfDrive= is_SelfDrive

selfDrive= Tesla(4, 4, "CNG", True)

selfDrive.drive()

```

```python
class Animal:
    def __init__(self, Name):
        self.Name = Name

    def Speak(self):
        print("subclass must implement  this method")

class Pet:
    def __init__(self, Owner):
        self.Owner=Owner

#drive class
class Dog(Animal , Pet):
    def __init__(self ,Name , Owner):
        Animal.__init__(self ,Name)
        Pet.__init__(self, Owner)

    def Speak(self):
        return f"{self.Name} say woof"
    
# create object 

dog = Dog("buddy", "ayush")
print(dog.Speak())
print(dog.Owner)
```
