Operator overloading allows you to define the behavior of operators +-* etc for custom object you achieve this by overriding specific magic methods in ypur class

 

```python
#
class vectors:
    def __init__(self, x,y):
        self.x=x
        self.y=y

    def __add__(self, other):
        return vectors(self.x+other.x, self.y+other.y)
    
    def __sub__(self, other):
        return vectors(self.x - other.x, self.y -other.y)
    
    def __mul__(self, other):
        return vectors(self.x * other.x, self.y * other.y)

    def __eq__(self, other):
        return vectors(self.x == other.x, self.y == other.y)

    def __repr__(self):
        return f"vector({self.x} , {self.y})"

v1 = vectors(2,3)
v2 = vectors(4,5)

print(v1 + v2)
print(v1 - v2)
print(v1 * v2)
print(v1 == v2)
```

```python
class ComplexNumber :
    def __init__(self , real , imag):
        self.real = real
        self.imag = imag
    
    def __add__(self, other):
        return ComplexNumber(self.real+other.real , self.imag +other.imag)
    
c1 = ComplexNumber(2,3)
c2 = ComplexNumber(1,4)

print(c1 + c2)
```
