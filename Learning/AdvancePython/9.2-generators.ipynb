Generator are the simple way to create iterator. they use the yield key word to produce a series of value lazily which means they generate value on the fly and do not store them in memory.

```python
def square(n):
    for i in range(n):
        yield i**2
```

```python
g=square(3)
print(g)
```

```python
next(g)
```

```python
for i in square(5):
    print(i)
```

```python
def test():
    yield 1
    yield 2
    yield 3
```

```python
t= test()
print(t)
```

```python
next(t)
```

```python
#Generator are help full for reading file data becoz it read line by line not entire file
    

def read_file(fileName):
    with open(fileName ,"r") as file:
        for line in file:
            yield line
```

```python
file_name = "Demo.txt"

for line in read_file(fileName=file_name):
    print(line)
```
