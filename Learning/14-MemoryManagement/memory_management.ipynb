# reference counting

Reference count is the primary method in python uses to manage memory. Each object in python maintains a count reference pointing to it when the reference count drops to zero the memory occupied by the object is deallocated

```python
import sys

a = []
## 2 one reference from the variable a and one from the argument of getrefcount
print(sys.getrefcount(a))
```

```python
b= a
print(sys.getrefcount(a))
```

```python
del b 
print(sys.getrefcount(a))
```

#### garbage collection
python includes a cyclic garbage collector that can detect 
and collect cyclic references.
A cyclic reference occurs when two or
more objects reference each other,
creating a cycle that prevents the reference 
count from reaching zero. The garbage collector periodically 
runs to identify and clean up these cycles, ensuring
that memory is freed even when reference counting alone
cannot determine when to delete objects.

```python
import gc
gc.enable()
```

```python
gc.disable()
```

```python
gc.collect()
```

```python
#get garbage collection static

print(gc.get_stats())
```

```python
## get unreachable objects

print(gc.garbage)
```

Memory Management Best Practices
1. Use Local Variables: loca variables have a shorter lifespan and are freed sooner then global variables.
2. Avoid Circular Reference : Circular reference can lead to memory leaks if not properly managed.
3. Use Generators: Generators produces items one at a time and only keep one time in memory at a time, making then memory efficient.
4. Explicitly delet objects :Use the del statement to delete variables and objects explicitly.
5. Profiel memory usage: Use memory profiling tools like tracemalloc and memory_profiler to identify memory leaks and optimize memory usage.

```python
import gc 
#handling circular reference
class myobject:
    def __init__(self, name):
        self.name = name
        print(f"object  {self.name} created")

    def __del__(self):
        print(f"object  {self.name} deleted")

#create circular reference
obj1 = myobject("obj1")
obj2 = myobject("obj2")
obj1.ref = obj2
obj2.ref = obj1

del obj1
del obj2


#manually trigger garbage collection
gc.collect()
```

```python
print(gc.garbage)
```

```python
## Generators for memory efficiency

# generator allow us to create iterators that generate values
#  on the fly instead of storing them in memory all at once.
#  This can be especially useful when working with large
#  datasets or infinite sequences.
def my_generator(n):
    for i in range(n):
        yield i 

#using the generator
for value in my_generator(10):
    print(value)
```

```python
## profiling Memory usage with tracemalloc

import tracemalloc

def create_list(n):
    return [i for i in range(n)]

def main():
    tracemalloc.start()
    create_list(1000000)
    current = tracemalloc.take_snapshot()
    top_stats = current.statistics('lineno')
    print("[ Top 10 ]")
    for stat in top_stats[:10]:
        print(stat)
```

```python
main()
```
